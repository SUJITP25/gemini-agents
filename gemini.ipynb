```python
#Langgraph (Code With Arohi)
# Node - Means Task and Edge Means Connection 
# State - Store the information of the nodes 
# State is like a memory and state machine is a decison maker 
# Branching logic, fallback meachanism, Multi-step decisions

from langgraph.graph import StateGraph,START,END,add_messages 
from langchain_community.tools import WikipediaQueryRun 
from langchain_community.utilities import WikipediaAPIWrapper
from typing import Dict,List
from langgraph.prebuilt import create_react_agent
from langchain_core.messages import HumanMessage

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key=GEMINI_API_KEY
)


api_wrapper = WikipediaAPIWrapper(top_k_results=1,doc_content_chars_max=100)
wiki_tool= WikipediaQueryRun(api_wrapper=api_wrapper)

# wiki_tool.run({"query": "AI Agents"})


# Custom Tools 


tools = [wiki_tool]
llm_with_tools = llm.bind_tools(tools)
result = llm_with_tools.invoke("Hello World")
result.content

agent_executor = create_react_agent(llm,tools)

response =agent_executor.invoke({'messages':[HumanMessage(content="What is Virat Kohli's Birthday?")]})
response["messages"][-1].content
```
